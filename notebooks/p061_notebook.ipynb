```python
def figurate_class(num, figurate_dict):

    for key in figurate_dict:
        if num in figurate_dict[key]:
            return key
```

```python
def initial_digits_list(first_num, l, figurate_dict, existing_classes):
    
    limited_list = []
    existing_classes = [figurate_class(first_num, figurate_dict)]

    for num in l:
        num_class = figurate_class(num, figurate_dict)
        if num_class not in existing_classes and initial_digits(num) == last_digits(first_num):
            limited_list.append(num)
            existing_classes.append(num_class)

    return limited_list, existing_classes
```

```python
def cyclical_check_list(cyclical_list):
    i = 0
    
    while i <= len(cyclical_list) - 2:
        x, y = [cyclical_list[i], cyclical_list[i+1]]
        if not cyclical_check(cyclical_list[-1], cyclical_list[0]) or not cyclical_check(x, y):
            return False
        i += 1

    return True
```
